```python
import numpy as np
import pandas as pd

##Importing dataset
dataset = pd.read_csv('../dataset/Diabetes.csv',encoding='utf-8')
```

```python
#obtaining numeric columns names
numerics_types = ['int32','int64','float16','float32','float64']
cols_num =  dataset.select_dtypes(include=numerics_types)

#selecting numeric atributes
numerics_columns = list(cols_num.columns)

##take target atribute
class_column = dataset['classe']

#splitting class atribute
x = dataset[numerics_columns] #features
y = dataset.classe            #target variable (class)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix

#spliting dataset into two parts: training and test
x_train_70, x_test_30, y_train_70, y_test_30 = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
#create a decision tree classifier object
clf = DecisionTreeClassifier(criterion='entropy',max_depth=3)

#Train Decision Tree Classifier
clf = clf.fit(x_train_70,y_train_70)

#predict the response for test dataset
y_pred = clf.predict(x_test_30)

#model accuracy
acuracia = metrics.accuracy_score(y_test_30,y_pred)
print('Accuracy: %.3f' % acuracia)

#confusion matrix for 30%
confusion_matrix(y_test_30,y_pred)

```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12)) #set plot size (denoted in inches)
tree.plot_tree(clf,fontsize=10)
plt.show()

#confusion matrix for  30%
confusion_matrix(y_test_30,y_pred)
```
