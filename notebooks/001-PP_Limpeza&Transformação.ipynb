```python
### Carregar as Libraries
import numpy as np
import pandas as pd

### Importing Dataset
dataset = pd.read_csv('Adult.csv',encoding='utf-8')

# Visualização dos atributos
dataset.info()
```

```python
missing = dataset.isnull().sum()
print(missing)
```

```python
## Deletando (apagando) todas as linhas com missing values
dataset2 = dataset.dropna()
missing = dataset2.isnull().sum()
print(missing)
```

```python
##Suavizando os missing values, ao invés de deletar e sofrer o provável desbalanceamento
##Importando a função para achar a moda
from scipy.stats import mode

##Alterando missing com a moda
dataset['workclass'].fillna(mode(dataset['workclass']).mode[0],inplace=True)
dataset['occupation'].fillna(mode(dataset['occupation']).mode[0], inplace=True)
dataset['native_country'].fillna(mode(dataset['native_country']).mode[0],inplace=True)

dataset.head(10)

```

```python
###  NORMALIZANDO ATRIBUTOS NUMÉRICOS ###

## Carregando Library para normalizar atributos numéricos
from sklearn.preprocessing import minmax_scale

# Obtendo os nomes das colunas Numéricas
tipos_numericos = ['int64']
cols_num = dataset.select_dtypes(include=tipos_numericos)

## selecionando as colunas
colunas_numericas = list(cols_num.columns)

# Copiando os dados e aplicando a normalizacao
dados_normalizados = dataset.copy()
dados_normalizados[colunas_numericas] = dataset[colunas_numericas].apply(minmax_scale)

# Visualização dos dados normalizados
dados_normalizados.head(10)
```

```python
## BINARIZAÇÃO DE ATRIBUTOS CATEGÓRICOS

## Pegar a classe
col_classe = dataset['class']

## Apagar a classe
dataset.drop('class',inplace=True,axis=1) #??????

#Obtendo os nomes das colunas categóricas
tipos_categoricos = ['object']
cols_cat = dataset.select_dtypes(include=tipos_categoricos)

# Binarizando os atributos categóricos
categoricoToBinary = pd.get_dummies(cols_cat) # pesquisar ?????
dados = dataset.join(categoricoToBinary)

# Dropando os atributos categóricos originais
dados.drop(cols_cat.columns.to_list(),axis=1,inplace=True)

# Juntando o atributo classe
final_data = dados.join(col_classe) # o que é esse class??????

#Visualização dos dados categóricos normalizados
final_data.head()
```

```python
##DISCRETIZANDO ATRIBUTOS NUMÉRICOS (PROCESSO INVERSO À TRANSFORMAÇÃO NOMINAL->NUMÉRICA)

#Ordena as idades
np.array(sorted(dataset.age.unique()))

# insere o atributo nominal com 3 categorias
dataset['age_nominal'] = pd.cut(dataset.age,bins=3,labels=['jovem','adulto','idoso']) ## Como isso é dividido (numericamente) internamente????

#Visualização dos dados normalizados (discretizados)

dataset.head()
```
